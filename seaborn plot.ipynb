```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(tips['total_bill'], bins=30, kde=True)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'])
```

```python
#
sns.barplot(x='sex', y='tip', data=tips, estimator=np.std)
```

```python
sns.countplot(x='size', data=tips)
```

```python
sns.jointplot(x='tip',y='size',data=tips, hue='sex')
```

```python
# {\displaystyle Q_{1}}Q_{1}25%
# {\displaystyle Q_{2}}Q_{2}50%
# {\displaystyle Q_{3}}Q_{3}75%
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# 
sns.boxplot(x='day',y='tip',data=tips, hue='smoker')
```

```python
# 
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

```python
#scatter plot,  non-numeric   
sns.stripplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
#
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
sns.swarmplot(x='day',y='total_bill',data=tips, hue='sex', color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind='bar')
```
