```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(tips['total_bill'], bins=30, kde=True)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'])
```

```python
#男性離散程度高。表示有時小費高低差異較大
sns.barplot(x='sex', y='tip', data=tips, estimator=np.std)
```

```python
sns.countplot(x='size', data=tips)
```

```python
sns.jointplot(x='tip',y='size',data=tips, hue='sex')
```

```python
# 第一四分位數（{\displaystyle Q_{1}}Q_{1}），又稱較小四分位數，等於該樣本中所有數值由小到大排列後第25%的數字。
# 第二四分位數（{\displaystyle Q_{2}}Q_{2}），又稱中位數，等於該樣本中所有數值由小到大排列後第50%的數字。
# 第三四分位數（{\displaystyle Q_{3}}Q_{3}），又稱較大四分位數，等於該樣本中所有數值由小到大排列後第75%的數字。
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#在星期五 抽菸者願意付的小費比較多
sns.boxplot(x='day',y='tip',data=tips, hue='smoker')
```

```python
#在星期五，星期六 抽菸者願意付的帳單比較多
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

```python
#可以出現scatter plot, 分類為 non-numeric 的資料，才有意義 如星期一 星期二
sns.stripplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
#點和點不會重疊
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
sns.swarmplot(x='day',y='total_bill',data=tips, hue='sex', color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind='bar')
```
